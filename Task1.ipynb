```python
import pandas as pd
```

```python
# Step 1: Read the Excel sheets into DataFrames
source_file = 'data bank0.xlsx'
target_file = 'import.xlsx'

source_df = pd.read_excel(source_file)
target_df = pd.read_excel(target_file)

# Step 2: Extract the 'Model Number' and 'Brand Name' columns from the source DataFrame
model_name = source_df['Model Name']
brand_name = source_df['Brand Name']

# Step 3: Add the columns to the target DataFrame
target_df['Model Name'] = model_name
target_df['Brand Name'] = brand_name

# Step 4: Write the updated DataFrame back to the Excel file
target_df.to_excel(target_file, index=False)
```

```python
updated_file = 'import.xlsx'
updated_df = pd.read_excel(updated_file)
merged_df = pd.merge(source_df, target_df, on='Product Description', how='inner')
merged_df
```

```python
# Remove the `column name' which contains NAN and also BLANK values
new_df = merged_df.drop(["Consignee Address 2","Raw Shipper Address2","Raw Consignee Phone",
                          "Shipper City","Shipper Country","Port Of Origin","Raw Consignee Add2","Shipper Address 2 "], axis=1)

# Save the DataFrame to a file
new_df.to_excel("data_new.xlsx")
new_df = pd.read_excel("data_new.xlsx")
new_df

```

```python
from difflib import SequenceMatcher

# Function to calculate percentage similarity between two strings using SequenceMatcher
def get_percentage_similarity(a, b):
    return SequenceMatcher(None, a, b).ratio() * 100

# Calculate the percentage match for each row in updated_df
percentage_match_list = []
for index, row in new_df.iterrows():
    import_description = row['Product Description']
    data_bank_description = row['Product Description']
    percentage_match = get_percentage_similarity(import_description, data_bank_description)
    percentage_match_list.append(percentage_match)

# Add the percentage match list as a new column to updated_df
new_df['Percentage_Match'] = percentage_match_list
new_df
```

```python
def get_percentage_similarity(a, b):
    return SequenceMatcher(None, a, b).ratio()*100
similarity_threshold = 1
matched_rows = new_df[new_df.apply(lambda row: get_percentage_similarity(row['Product Description'], 
                                row['Product Description']) >= similarity_threshold, axis=1)]
result_df = matched_rows[['Model Name_y', 'Brand Name_y']].rename(columns={'Brand_Name_y': 'Brand Name', 'Model_Name_y': 'Model Name'})
result_df
```
